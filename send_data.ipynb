```python

import json
import requests
import os
import time
import pandas as pd
from typing import Dict

def send_data(df):
    max_index = max(0, df.shape[0])

    for idx in range(100*max_index):
        dataset_size = df.shape[0]
        data = df.iloc[idx % dataset_size].to_dict()
        requests.post(
            "http://3.80.42.207:5002/prediction/",
            data=json.dumps(data),
            headers={"content-type": "application/json"},
        )
        if idx % 20 == 0:
            print("Wait 5 seconds till the next try.")
            time.sleep(5)
    
```

```python

df = pd.read_csv('data/prod.csv').iloc[:, :-3]
import numpy as np
data_drift = df.copy()
mu, sigma = 3, 1
noise = np.random.normal(mu, sigma, [980,11])
noise[:, -1] =0 
data_drift = data_drift + noise
```

```python
send_data(df)
```

```python

send_data(data_drift)
```
